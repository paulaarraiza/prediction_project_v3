## **This notebook aims to compare datatypes and generate visual statistics for them**

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
os.environ["CUDA_VISIBLE_DEVICES"] = "0"  # Use GPU 0 in first notebook
```

```python
project_dir = "/home/jupyter-tfg2425paula/prediction_project_v3"
os.chdir(project_dir)

results_dir = os.path.join(project_dir, "02_results")
plots_dir = os.path.join(project_dir, "03_plots")
```

```python
lstm_results = pd.read_csv(os.path.join(results_dir, 'individual_trials/lstm_pca_accuracy_bydatatype.csv'))
gru_results = pd.read_csv(os.path.join(results_dir, 'individual_trials/gru_pca_accuracy_bydatatype.csv'))

lstm_summarystats = pd.read_csv(os.path.join(results_dir, 'individual_trials/lstm_pca_bydatatype_summarystats.csv'))
gru_summarystats = pd.read_csv(os.path.join(results_dir, 'individual_trials/gru_pca_bydatatype_summarystats.csv'))
```

```python
summary_stats = pd.concat([lstm_summarystats, gru_summarystats])

# Set seaborn style
sns.set_style("whitegrid")

# Plot Mean Accuracy per Data Type
plt.figure(figsize=(10, 5))
sns.barplot(data=summary_stats, x="DATA_TYPE", y="mean", hue="MODEL", palette="viridis")
plt.title("Mean Accuracy by Data Type for LSTM & GRU")
plt.xlabel("Data Type")
plt.ylabel("Mean Accuracy")
plt.legend(title="Model")
plt.show()

# Plot Standard Deviation of Accuracy per Data Type
plt.figure(figsize=(10, 5))
sns.barplot(data=summary_stats, x="DATA_TYPE", y="std", hue="MODEL", palette="coolwarm")
plt.title("Standard Deviation of Accuracy by Data Type for LSTM & GRU")
plt.xlabel("Data Type")
plt.ylabel("Standard Deviation")
plt.legend(title="Model")
plt.show()

# Boxplot of Accuracy per Data Type
plt.figure(figsize=(10, 5))
sns.boxplot(data=summary_stats, x="DATA_TYPE", y="mean", hue="MODEL", palette="pastel")
plt.title("Boxplot of Accuracy by Data Type for LSTM & GRU")
plt.xlabel("Data Type")
plt.ylabel("Accuracy")
plt.legend(title="Model")
plt.show()
```

```python
summary_stats.to_csv(os.path.join(results_dir, 'individual_trials/pca_summary_stats.csv'), index=False)
```
