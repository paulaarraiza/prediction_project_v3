```python
import pandas as pd
import os
```

```python
project_dir = "/home/jupyter-tfg2425paula/prediction_project_v3"
os.chdir(project_dir)

raw_data_dir = os.path.join(project_dir, "00_data")
options_data_dir = os.path.join(raw_data_dir, "raw/options")

stocks = ['AAPL', 'MSFT', 'AMZN', 'NVDA', 'SPX']
```

```python
df = pd.read_csv(os.path.join(options_data_dir, "combined_options.csv"))
df
```

```python
for stock in stocks:
    # Select columns starting with the stock name plus the 'Date' column
    filtered_df = df[['Date'] + [col for col in df.columns if col.startswith(stock)]]

    # Define file name
    output_filename = f"{stock}_options.csv"

    # Save the filtered DataFrame
    filtered_df.to_csv(os.path.join(options_data_dir, output_filename), index=False)

    print(f"Saved {output_filename}")
```
